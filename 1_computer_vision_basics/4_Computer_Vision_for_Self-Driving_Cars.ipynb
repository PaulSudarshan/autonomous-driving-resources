### Sharpening and blurring

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('test_image.jpg')
cv2.imshow('My Image', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Converting into Grayscale
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('Self Driving Car in Gray!', gray_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Defining Kernal
Sharp_Kernel_1 = np.array([[0,-1,0], 
                        [-1,5,-1], 
                        [0,-1,0]])

#Applying Kernal

Sharpened_Image_1 = cv2.filter2D(gray_img, -1, Sharp_Kernel_1)
```

```python
cv2.imshow('Sharpened Image', Sharpened_Image_1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Sharpening Kernal 2
Sharp_Kernel_2 = np.array([[-1,-1,-1], 
                        [-1,9,-1], 
                        [-1,-1,-1]])
```

```python
Sharpened_Image_2 = cv2.filter2D(gray_img, -1, Sharp_Kernel_2)
```

```python
cv2.imshow('Sharpened Image_2', Sharpened_Image_2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Blurring Kernal

Blurr_Kernel = np.ones((3,3))
Blurr_Kernel
```

```python
Blurred_Image = cv2.filter2D(gray_img, -1, Blurr_Kernel)
cv2.imshow('Blurred Image', Blurred_Image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Blurring Kernel normalized 
Blurr_Kernel = np.ones((3,3)) * 1/9
Blurr_Kernel

Blurred_Image = cv2.filter2D(gray_img, -1, Blurr_Kernel)
cv2.imshow('Blurred Image', Blurred_Image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
Blurr_Kernel = np.ones((8,8))
Blurr_Kernel
```

```python
#output of blurr kernal
Blurr_Kernel = np.ones((8,8)) * 1/64
Blurr_Kernel
```

```python
#Checking the blurred image 
Blurred_Image = cv2.filter2D(gray_img, -1, Blurr_Kernel)

cv2.imshow('Blurred Image', Blurred_Image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Canny edge detection

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('test_image.jpg')

cv2.imshow('My Test Image', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('Self Driving Car in Gray!', gray_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Sobel X
x_sobel = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1, ksize = 7)
cv2.imshow('Sobel - X direction', x_sobel)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Sobel Y
y_sobel = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize = 7)
cv2.imshow('Sobel - Y direction', y_sobel)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Laplacian
laplacian = cv2.Laplacian(gray_img, cv2.CV_64F)
cv2.imshow('Laplacian', laplacian)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Canny Edge detection 

threshold_1 = 120
threshold_2 = 200

canny = cv2.Canny(gray_img, threshold_1, threshold_2)

cv2.imshow('Canny', canny)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Perspective Transform

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
height
```

```python
width
```

```python
Source_points = np.float32([[420,70], [580, 50], [590,210], [430, 220] ])
```

```python
Destination_points = np.float32([[0,0], [width,0], [width,height], [0,height]])
```

### Erosion, Input, Dilation

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('Speed_Sign_View_2.jpg')
cv2.imshow('Original Image', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Regions of interest masking

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image_color = cv2.imread('test5.jpg')
cv2.imshow('Original Image', image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
height, width = image_color.shape[:2]
```

```python
height
```

```python
width
```

```python
# Converting image in grayscale
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_gray, cmap = 'gray')
```

```python
#Point of interest in the image
ROI = np.array([[(0, 400),(300, 250), (450, 300), (700, height)]], dtype=np.int32)
```

```python
#Define blank image with all zeros
blank = np.zeros_like(image_gray)
```

```python
blank.shape
```

```python
#Filling mask of interest
mask = cv2.fillPoly(blank, ROI, 255)
```

```python
#Perform a bit-wise AND operation
masked_image = cv2.bitwise_and(image_gray, mask)
```

```python
#Masked Image
plt.imshow(masked_image, cmap = 'gray')
```

### Hough transform

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_c = cv2.imread('calendar.jpg')
cv2.imshow('Given Image', image_c)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image_g = cv2.cvtColor(image_c, cv2.COLOR_BGR2GRAY)
```

```python
image_canny = cv2.Canny(image_g, 50, 200, apertureSize= 3)
```

```python
image_canny
```

```python
cv2.imshow('canny image', image_canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Applying Hough transform
lines = cv2.HoughLines(image_canny, 1, np.pi/180, 250)
```

```python
 if lines is not None:
        for i in range(0, len(lines)):
            rho = lines[i][0][0]
            theta = lines[i][0][1]
            
            x0 = rho * np.cos(theta)
            y0 = rho * np.sin(theta)
            
            a = np.cos(theta)
            b = np.sin(theta)
           
            x1 = int(x0 + 1000 * (-b))
            y1 = int(y0 + 1000 * (a))
            x2 = int(x0 - 1000 * (-b))
            y2 = int(y0 - 1000 * (a))
            cv2.line(image_c, (x1, y1), (x2, y2), (255, 0, 0), 2)
```

```python
cv2.imshow('Hough Lines', image_c)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
